```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
import os
from os import listdir
```

```
target=[]
flat_data=[]
images=[]
dir="/content/drive/MyDrive/ML/images"
cat=["Bikes","Cars"]

for i in cat:
  y=cat.index(i)
  path=os.path.join(dir,i)
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    img_resized=resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(y)
flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
```

```
data=pd.DataFrame(flat_data)
data["target"]=target
data
```

```
x=pd.DataFrame(flat_data)
y=data["target"]
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix,classification_report
lr=LogisticRegression()
dtc=DecisionTreeClassifier()
gnb=GaussianNB()
mnb=MultinomialNB()
bnb=BernoulliNB()
knc=KNeighborsClassifier()
svc=SVC()
rfc=RandomForestClassifier()
xg=XGBClassifier()
```

**Here we can see that gnb and svc has performed better than other algorithms .so, lets try to improve them using hyperparamater tuning.**

```
param_grid = {'C': [0.1, 1, 10, 100, 1000],'gamma': [1, 0.1, 0.01, 0.001, 0.0001]}  
  
grid = GridSearchCV(svc, param_grid, refit = True, verbose = 3,n_jobs=10) 
grid.fit(x_train, y_train) 
```

```
grid.best_params_
```

```
svcb=SVC(C=10,gamma=0.0001)
```

```
svcb.fit(x_train,y_train)
```

```
svcb_pred=svcb.predict(x_test)
```

```
accuracy_score(svcb_pred,y_test)
```

```
confusion_matrix(svcb_pred,y_test)
```

```
classification_report(svcb_pred,y_test)
```

**here we can see that there is no improvement in the hyperparameter tuning of SVC algo**
